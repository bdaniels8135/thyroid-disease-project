# Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Read in Data

```python
X_train = pd.read_csv('train.csv')
y_train = pd.read_csv('train_labels.csv')
```

# Normalize Data

```python
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
```

# Run PCA

```python
pca = PCA()
X_train_pca = pca.fit_transform(X_train_scaled)
pca.explained_variance_ratio_
```

```python
plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=np.array(y_train))
plt.show()
```

```python
plt.scatter(X_train_pca[:,1], X_train_pca[:,2], c=np.array(y_train))
plt.show()
```

```python
plt.scatter(X_train_pca[:,0], X_train_pca[:,2], c=np.array(y_train))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X_train_pca[:,0], X_train_pca[:,1], X_train_pca[:,2], c=np.array(y_train))
plt.show()
```

# Conclusions

PCA did not work particularly well with this data set since it takes many components to accumulate significant explained variance.
